```python
ruta_datos = '/content/drive/My Drive/reporte_mercado_libre.xlsx'
```

```python
df = pd.read_excel(ruta_datos)
```

```python
df.columns
```

```python
print({len(df )})
```

```python
# Eliminar las columnas que no se necesitan
df = df.drop (columns=['Canal de Venta', 'Cliente', 'Provincia',
       'Cantidad vendida', 'Precio unitario', 'Precio unitario dolar',
       'Valor MEP', 'Número de publicación',
       'Categoría de la publicación'])
```

```python
# Agrupar total por venta para cada día
df['Total acumulado'] = df.groupby('Fecha de venta')['Total venta dolar'].transform('sum')
```

```python
df.head()
```

```python
# Eliminar la columna Total venta dolar
df = df.drop (columns=['Total venta dolar'])

df.head()
```

```python
# Encontrar filas repetidas basadas en la columna 'Fecha de venta'
repetidas = df[df.duplicated(subset=['Fecha de venta'], keep=False)]

# Mostrar las filas duplicadas
print("Filas repetidas:")
print(repetidas)
```

```python
# Eliminar filas con iguales valores, conservando solo la primera ocurrencia de cada 'Fecha de venta'
df_sin_repetidos = df.drop_duplicates(subset='Fecha de venta', keep='first')
```

```python
print({len(df_sin_repetidos )})
```

```python
print(df_sin_repetidos)
```

```python
df_sin_repetidos = df_sin_repetidos.sort_values(by='Fecha de venta', ascending=True)
```

```python
print(df_sin_repetidos)
```

```python
df_sin_repetidos = df_sin_repetidos.rename(columns={'Fecha de venta': 'Fecha', 'Total acumulado': 'Acumulado' })
```

```python
print(df_sin_repetidos)
```

```python
# Establecer la columna Fecha de venta como índice
df_sin_repetidos.set_index('Fecha', inplace=True)
```

```python
print(df_sin_repetidos)
```

```python
# Guardar el df_sin_repetidos en un nuevo archivo en el drive, para luego trabajar en los modelos

df_sin_repetidos.to_excel('/content/drive/My Drive/reporte_para_serie_temporal.xlsx', index=True)
```
